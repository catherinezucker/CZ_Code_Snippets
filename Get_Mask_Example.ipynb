# Now, enter contour level you want to use to define the mask. You will also need to enter the pixel coordinate of a point INSIDE this contour

```python
#Set the contour level threshold you want to use to define the filament
contour_level=1.25e+22

#Set the name of the input file (in our case, the cropped image)
input_filestr="HIGAL_L_Fil8_N.fits"

#Set the name of what you want to call the output file
output_filestr="higal_mask.fits"

fig, axes = plt.subplots(1)

#Read in image data
image_data,header=fits.getdata(input_filestr,header=True)

#You made need to smooth it a bit to apply a contour (check out the gaussian filter documentation):
image_data_smoothed=ndimage.gaussian_filter(image_data, sigma=(2.,2.))

#Show intensity image
axes.imshow(image_data_smoothed,origin='lower',cmap='Greys',norm=LogNorm(vmin=4.2e+21,vmax=3e+22))

#Plot contours
c=axes.contour(image_data_smoothed,levels=[contour_level],colors=['lime'])

#Now enter a coordinate that corresponds to a point inside your mask; 
#We need this so we can choose which contour to make a mask from, in case there are many contours in an image. 
infilcoords=[[184,153]]

#Plot it to make sure it is actually inside! I have overlaid a point in red as an example
plt.scatter(infilcoords[0][0],infilcoords[0][1],c='r')
```

# Now we will make a mask based on the contour you specified, and write it out to a file!

```python
#Convert contours to list of points
p = c.collections[0].get_paths()

#Extract point(s) inside polygon(s) corresponding to chosen contour
points=list(infilcoords)

#Now collect pixels inside it
for (point,filnum) in zip(points,range(0,len(points))):

    index=[]
    for i in range(0,len(p)):
        if p[i].contains_point((point[0],point[1]))==True:
            index.append(i)

    print(index)
    filpoints=np.empty((0,2))
    xpix=[]
    ypix=[]
    for k in range(0,len(index)):              
        x=p[index[k]].vertices[:,0]
        y=p[index[k]].vertices[:,1]

        poly=Polygon([(i[0], i[1]) for i in zip(x,y)])
        xpix.append([i for i in x])
        ypix.append([i for i in y])

        patch=PolygonPatch(poly)

        # calculate the x and y points possibly within the image
        y_int = np.arange(0, image_data_smoothed.shape[0])
        x_int = np.arange(0, image_data_smoothed.shape[1])

        # create a list of possible coordinates
        g = np.meshgrid(x_int, y_int)
        coords = list(zip(*(c.flat for c in g)))

        # create the list of valid coordinates inside contours
        filpoints = np.vstack((filpoints,np.vstack([p for p in coords if patch.contains_point(p, radius=0)])))
        
#Make an empty image set points inside mask to 1
empty=np.zeros((image_data_smoothed.shape))
empty[filpoints[:,1].astype(int),filpoints[:,0].astype(int)]=1

#Display mask
fig=plt.figure()
plt.imshow(empty,origin='lower',cmap='binary')

#Output mask to file
fits.writeto(output_filestr,empty,header,overwrite=True)
```
